```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('dataset\\Family Income and Expenditure.csv')
```

<h3>Completing missing data</h3>

```python
df.isnull().sum()
```

```python
df['Household Head Occupation'] = df['Household Head Occupation'].fillna("Unspecified")
df['Household Head Class of Worker'] = df['Household Head Class of Worker'].fillna("Unspecified")
df['Toilet Facilities'] = df['Toilet Facilities'].fillna("Unspecified")
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Region'] =df['Region'].astype('category')
df['Main Source of Income'] = df['Main Source of Income'].astype('category')
df['Household Head Sex'] = df['Household Head Sex'].astype('category')
df['Household Head Marital Status'] =df['Household Head Marital Status'].astype('category')
df['Household Head Job or Business Indicator'] = df['Household Head Job or Business Indicator'].astype('category')
df['Household Head Occupation'] = df['Household Head Occupation'].astype('category')
df['Household Head Class of Worker'] = df['Household Head Class of Worker'].astype('category')
df['Type of Household'] = df['Type of Household'].astype('category')
df['Household Head Occupation'] = df['Household Head Occupation'].astype('category')
df['Type of Building/House'] = df['Type of Building/House'].astype('category')
df['Type of Roof'] = df['Type of Roof'].astype('category')
df['Tenure Status'] = df['Tenure Status'].astype('category')
df['Toilet Facilities'] = df['Toilet Facilities'].astype('category')
df['Main Source of Water Supply'] = df['Main Source of Water Supply'].astype('category')
```

<h1>Average Number of Family Members</h1>
<h3>The average number of family members per household is 4.6, offering valuable insight into the typical family structure within the community.

Larger households often face unique financial and social challenges, including higher living expenses and increased demand for basic services like healthcare, education, and housing. Understanding household size is essential for effective community planning, allowing social services and local governments to better tailor their programs and allocate resources to meet the needs of families.

</h3>

```python
average_family_members = df['Total Number of Family members'].mean()
print(f'Average Number of Family Members: {average_family_members}')
```

<h1>Average Number of Family Members per Region</h1>
<h3>The Autonomous Region in Muslim Mindanao (ARMM) has the highest average family size at 5.71, indicating a tendency toward larger households, possibly due to cultural or socio-economic factors. In contrast, regions like Cagayan Valley, MIMAROPA, and Davao have smaller averages, suggesting shifts toward nuclear families or better access to family planning. Urbanized regions such as NCR maintain moderate family sizes, reflecting the influence of urban living on household composition.</h3>

```python
average_family_members_per_region = df.groupby('Region')['Total Number of Family members'].mean()

# Display the result
print("Average Number of Family Members per Region:")
print(average_family_members_per_region)
```

```python
average_family_members_per_region.plot(kind='barh', color='skyblue')
plt.title('Average Family Size per Region')
plt.xlabel('Average Family Members')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

<h1>Are rural families consistently larger than urban ones?</h1>
<h3>The analysis shows that rural families have a larger average number of family members compared to urban families. This trend is consistent with demographic patterns where rural areas tend to have bigger households, possibly due to cultural preferences for extended families, reliance on agriculture or manual labor (which benefits from larger households), and limited access to family planning resources. In contrast, urban families tend to be smaller, likely influenced by higher living costs, space limitations, and lifestyle changes associated with urban living.</h3>

```python
df_filtered = df[df['Type of Household'].notna()]
plt.figure(figsize=(8, 5))
sns.barplot(
    x='Type of Household',
    y='Total Number of Family members',
    data=df_filtered,
    ci=None,
    palette='Set2'
)
plt.title('Average Family Size: Urban vs Rural')
plt.xlabel('Type of Household')
plt.ylabel('Average Number of Family Members')
plt.tight_layout()
plt.show()
```

<h1>Housing Affordability </h1>
<h3>The Housing Affordability Index reflects the percentage of household income spent on housing costsâ€”the lower the percentage, the more affordable housing is for residents.

The National Capital Region (NCR) has the lowest housing affordability index at 8.24%, indicating that residents spend a smaller share of their income on housing, likely due to higher incomes despite higher housing costs.

On the other hand, regions like Zamboanga Peninsula (17.21%), Cagayan Valley (16.33%), and MIMAROPA (15.49%) show the highest percentages, signaling greater financial pressure on households for housing expenses.

Regions with higher housing affordability indices may benefit from affordable housing programs, subsidies, or housing policy reforms to reduce the burden on residents.</h3>

```python
affordability =df.groupby('Region').apply(lambda x: x['Total Household Income'].mean() / x['Imputed House Rental Value'].mean())
affordability = affordability.reset_index()
print(affordability)
    
```

<h1> Regions with higher income also show higher restaurant/hotel expenditures.</h1>
<h3>The National Capital Region (NCR) shows a stark contrast in total household income compared to other regions, reaching over PHP 400,000, highlighting its economic primacy.

Low Restaurant Spending: Despite the high income in NCR, restaurant and hotel expenditures are markedly low across all regions, with minimal spending compared to total income.

<h1>Among food types, rice and meat have the strongest correlation with total food spending.</h1>
<h3>The highest correlation is observed between Meat Expenditure and Total Food Expenditure, with a value of 0.77, indicating a strong relationship. Additionally, Total Fish and Marine Products Expenditure shows a moderate correlation with both Meat Expenditure (0.47) and Total Food Expenditure (0.61). Other categories such as Fruit and Vegetables Expenditure demonstrate weaker correlations, particularly with Total Rice Expenditure, suggesting that spending on rice has less impact on overall food expenditure compared to meat and fish. This analysis can help in understanding spending behaviors and potential areas for cost management.</h3>

```python
food_cols = ['Total Food Expenditure', 'Total Rice Expenditure', 'Meat Expenditure',
             'Total Fish and  marine products Expenditure', 'Fruit Expenditure',
             'Vegetables Expenditure']
sns.heatmap(df[food_cols].corr(), annot=True, cmap='YlGnBu')
plt.title('Correlation Among Food Expenditures')
plt.show()
```
